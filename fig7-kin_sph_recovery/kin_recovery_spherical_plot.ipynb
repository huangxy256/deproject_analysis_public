# Plot the marginal PDF of the velocity dispersion ratio of spherical JAM and axisymmetric JAM

```python
import numpy as np
import matplotlib.pyplot as plt 
from copy import deepcopy
import h5py

from cycler import cycler
import matplotlib as mpl
clist = ['#04B2D9', '#F2B33D', '#F29966', '#F2441D', '#0486C5']
mpl.rcParams['axes.prop_cycle'] = cycler(color=clist)
```

```python
data_oblate = np.load('../fig5-fig6-make_kin_mock_data/kin_mock_data_oblate.npy')
sigma_e_oblate = data_oblate[2, :]
data_prolate = np.load('../fig5-fig6-make_kin_mock_data/kin_mock_data_prolate.npy')
sigma_e_prolate = data_prolate[2, :]

kin_recovery_sph_oblate = np.load('./kin_recovery_spherical_oblate.npy')
kin_recovery_sph_prolate = np.load('./kin_recovery_spherical_prolate.npy')

fig, ax = plt.subplots()
ax.set_box_aspect(0.7)
ax.hist(kin_recovery_sph_oblate/sigma_e_oblate, density=True, lw=2, histtype='step', label = 'oblate', bins=20, color=clist[0])
ax.hist(kin_recovery_sph_prolate/sigma_e_prolate, density=True, lw=2, histtype='step', label = 'prolate', bins=20, color=clist[2])
ax.axvline(np.mean(kin_recovery_sph_oblate/sigma_e_oblate), lw=2, ls='--', label = 'oblate mean', color=clist[0])
ax.axvline(np.mean(kin_recovery_sph_prolate/sigma_e_prolate), lw=2, ls='--', label = 'prolate mean', color=clist[2])
ax.set_xlabel(r'$\sigma_\mathrm{sph}^P/\sigma_\mathrm{axi}^P$', size = 18)
ax.set_ylabel('PDF', size = 18)
ax.tick_params(labelsize = 14)
# ax.set_xticks(np.arange(0.98, 1.06, 0.01))
ax.legend(prop = {'size': 14})
fig.savefig('../plots/kin_recovery_sph.pdf', dpi=300, bbox_inches='tight')
fig.show()
```
